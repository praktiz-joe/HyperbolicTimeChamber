# Recursion

## 1.1 Definitions
Recursion is a program that calls itself and it's used to solve problems by dividing them into smaller subproblems.
> You could use recursive functions to solve recurrence relation problems
 
## 1.2 Types of Recursion
### 1.2.1 Head Recursion
Tail recursion is when the last operation in the function/program is the recursive call.

```
def fun(n):
    {
        ...(statements)
    }
    fun(recursive call)
    
```
> Tail recursion is similar to a while loop. Usually in terms of complexity, both are O(N) but in space complexity, recursive calls are O(N) vs loops at O(1).

### 1.2.2 Tail Recursion
Head recursion is when the beginning operation after is the recursive call.

```
def fun(n):
    {
        if n > 0:
            fun(recursive call)
            ... (statements)
    }
```
> In the above example, the outputs are different from the loop output. You may have to modify the conditionals to get the results as in a loop

### 1.2.3 Tree Recursion
Tree recursion is when a recursive funcion/programs call itself more than one time in the code function.

```
def fun(n):
    {
        if n> 0: 
        {
            print(n)
            fun(n-1) # recursive call 1
            fun(n-1) # recursive call 2
        }
    }
```
> When you output the recursive calls in a diagram form, you'll se the tree form that it makes with levels and nodes. Usually, this kind of recursive calls are **O(2<sup>N</sup>)**, in terms of time complexity.

### 1.2.4 Indirect Recursion
Indirect recursion is more than one function/program calling recursively

```
def funA(n):
    {
        if n>0:
            print(n)
            funB(n-1)
    }
def funB(n):
    {
        if n>1:
            print(n)
            funA(n/2)
    }
```
> As you can see, fun A calls funB and then funB calls funA recursively.

### 1.2.5 Nested Recursion
Nested Recursion is when inside the recursive call, there's another recursive call being made

```
def fun(n):
    {
        if n> 100:
            return n -10
        else:
            return fun(fun(n+11))
    }
```
> The time complexity usually is **O(N<sup>2</sup>)**.
